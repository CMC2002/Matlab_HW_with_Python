# Problem 2-1

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage
from PIL import Image
```

```python
# Plot the result
def plot(data, title):
    plot.i += 1
    plt.subplot(2, 2, plot.i)
    plt.imshow(data)
    plt.gray()
    plt.title(title)
plot.i = 0
```

```python
# Loading data
im = Image.open('lena_gray.png')
data = np.array(im, dtype= float)
plot(data, 'Original')
```

```python
# Simple 3x3 Highpass
kernel = np.array([[-1, -1, -1],
                   [-1,  8, -1],
                   [-1, -1, -1]])
hp_3x3 = ndimage.convolve(data, kernel)
plot(hp_3x3, 'Simple 3x3 Highpass')
```

```python
# Simple 5x5 highpass filter
kernel = np.array([[-1, -1, -1, -1, -1],
                   [-1,  1,  2,  1, -1],
                   [-1,  2,  4,  2, -1],
                   [-1,  1,  2,  1, -1],
                   [-1, -1, -1, -1, -1]])
hp_5x5 = ndimage.convolve(data, kernel)
plot(hp_5x5, 'Simple 5x5 Highpass')
```

由上面兩圖比較可發現5x5的Highpass filter可以讓影像線條更明顯、更銳利化  
也可發現Higher highpass filter可以讓線條更明顯  
<b>符合Lab 1的結論</b>

```python
# plot function
import matplotlib.pyplot as plt
def plotfunction(x):
    plt.xlabel('t')
    plt.ylabel('x(t)')
    plt.plot(range(0, np.size(x)), x)
    plt.show()
```

Define 兩個 sin wave  
分別為 <i>sin(8x) sin(16x)</i> 

```python
freq1 = 8
freq2 = 16
dt = np.pi/100

x = np.arange(0, 2*np.pi, step= dt, dtype= float )
sin1 = np.sin(freq1 * x)
sin2 = np.sin(freq2 * x)
plotfunction(sin1)
plotfunction(sin2)
```

```python
# plot FFT
import matplotlib.pyplot as plt
def plotFFT(x):
    plt.xlabel('w')
    plt.ylabel('x(w)')
    plt.plot(range(0, np.size(x)), x)
    plt.show()
```

```python
from numpy.fft import fft
from numpy.fft import fftfreq
import matplotlib.pyplot as plt
Y1 = abs(fft(sin1))
Y2 = abs(fft(sin2))
w1 = fftfreq(np.size(sin1))*2*np.pi/dt
w2 = fftfreq(np.size(sin2))*2*np.pi/dt
plt.plot(w1, Y1, color = 'b')
plt.plot(w2, Y2, color = 'r')
plt.show()
```

藍色的為<i>sin(8x)</i>的<i>Fourier Transform</i>  
紅色的則為<i>sin(16x)</i>的 <i>Fourier Transform</i>  
且impulse的位置大約也等於他們對應的數字
