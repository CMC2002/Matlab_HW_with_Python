# Problem 1-1-a

```python
# Define function
import numpy as np
N = int(1001)
N1 = int((N-1)/2)
T = int(100)
Ts = T/N

n = np.arange(-N1, N1+1, step= 1, dtype= int)
x_n = np.zeros(N)
for i in range(0, N):
    if n[i]!=0:
        x_n[i] = np.sin(np.pi * Ts * n[i])/(np.pi * n[i] * Ts)
    else:
        x_n[i] = 1

x_n2 = np.zeros(201)
j = 0
for i in range(int(len(x_n)/2)-100,int(len(x_n)/2)+101):
    x_n2[j] = x_n[i]
    j+=1

x_n3 = np.zeros(101)
j = 0
for i in range(int(len(x_n)/2)-50,int(len(x_n)/2)+51):
    x_n3[j] = x_n[i]
    j+=1
```

```python
# plot function
import matplotlib.pyplot as plt
def plotfunction(x):
    plt.xlabel('n')
    plt.ylabel('x[n]')
    plt.stem(range(-int(np.size(x)/2), int(np.size(x)/2)+1), x)
    plt.show()
```

```python
# Problem 1-1-a
plotfunction(x_n)
plotfunction(x_n2)
plotfunction(x_n3)
```

# Problem 1-1-b

```python
# Discrete Time Fourier Transform
def DFT(x):
    p = np.size(x)
    n = np.arange(-int(p/2), int(p/2)+1, step= 1, dtype= int)
    k = n.reshape((p, 1))
    base = np.exp(-2j * np.pi *k *n/N)
    X = np.dot(base, x)
    return X
```

```python
# Plot DFT
def plotDFT(X):
    step = len(X)
    freq = np.arange(-int(step/2), int(step/2+1), step= 1, dtype= int)
    plt.figure(figsize= (8,6))
    plt.plot(freq, np.real(X))
    plt.xlabel('Freq(Hz)')
    plt.ylabel('DFT Amplitude X(freq)')
    plt.show()
```

```python
# Problem 1-1-b
X = np.real(DFT(x_n))
plotDFT(X)

X_n2 = np.zeros(201)
j = 0
for i in range(int(len(X)/2)-100,int(len(X)/2)+101):
    X_n2[j] = X[i]
    j+=1
plotDFT(X_n2)

'''
X_n3 = np.zeros(101)
j = 0
for i in range(int(len(X)/2)-50,int(len(X)/2)+51):
    X_n3[j] = X[i]
    j+=1
plotDFT(X_n3)
'''
```

```python
# Comparesion of different graph
def comparsion_DFT(X1, X2):
    if len(X1) != len(X2):
        print("Different shape. Failed")
    step = len(X1)
    freq = np.arange(-int(step/2), int(step/2+1), step= 1, dtype= int)
    plt.figure(figsize= (8,6))
    plt.plot(freq, X1)
    plt.plot(freq, X2)
    plt.xlabel('Freq(Hz)')
    plt.ylabel('DFT Amplitude X(freq)')
    plt.show()
```
